3. เขียนโปรแกรม infinity while loop โดยรับค่า input มาเรื่อยๆ จนกว่าผู้ใช้จะใส่เลข 0 และนำเลขทั้งหมดมาหาค่าเฉลี่ย


```python
total = 0
count = 0
value = float(input("กรุณาป้อนค่าตัวเลข: "))

while (value == 0):
    value = float(input("กรุณาป้อนค่าตัวเลข: "))
    count += 1
    
while (value != 0):
    total += value
    count += 1
    value = float(input("กรุณาป้อนค่าตัวเลข: "))
    
print(total); print(count)
average = total / count
print("ค่าเฉลี่ยคือ {}".format(average))
```

5. เขียนโปรแกรมรับค่า int มา 3 จำนวน และตรวจสอบว่า 3 เลขนั้นสามารถนำมาสร้าง 3 เหลี่ยมมุมฉากได้หรือไม่


```python
value1 = int(input("กรุณาป้อนจำนวนเต็มตัวที่ 1: "))
value2 = int(input("กรุณาป้อนจำนวนเต็มตัวที่ 2: "))
value3 = int(input("กรุณาป้อนจำนวนเต็มตัวที่ 3: "))

lst = [value1 ,value2 ,value3]
max_value = max(lst)
lst.remove(max_value)

if (sum(lst) < max_value):
    print("สร้างสามเหลี่ยมไม่ได้นะจ๊ะ")
else:
    if (lst[0]**2 + lst[1]**2 == max_value**2):
        print("สร้างสามเหลี่ยมมุมฉากได้")
    else:
        print("สร้างสามเหลี่ยมมุมฉากไม่ได้")
```

6. เขียนโปรแกรมรับค่า int มา 3 จำนวน และตรวจสอบว่าเมื่อนำ 3 เลขนั้นมาสร้าง 3 เหลี่ยม จะได้สามเหลี่ยมแบบใด
	
	- สามเหลี่ยมมุมแหลม

	- สามเหลี่ยมมุมป้าน

	- สามเหลี่ยมมุมฉาก


```python
value1 = int(input("กรุณาป้อนจำนวนเต็มตัวที่ 1: "))
value2 = int(input("กรุณาป้อนจำนวนเต็มตัวที่ 2: "))
value3 = int(input("กรุณาป้อนจำนวนเต็มตัวที่ 3: "))

lst = [value1 ,value2 ,value3]
max_value = max(lst)
lst.remove(max_value)

if (sum(lst) <= max_value):
    print("สร้างสามเหลี่ยมไม่ได้นะจ๊ะ")
else:
    if (lst[0]**2 + lst[1]**2 == max_value**2):
        print("สามเหลี่ยมมุมฉาก")
    elif (lst[0]**2 + lst[1]**2 > max_value**2):
        print("สามเหลี่ยมมุมแหลม")
    else:
        print("สามเหลี่ยมมุมป้าน")
```

7. เขียนโปรแกรมให้รับค่าอาหาร 3 ครั้งที่ไปทานอาหารนอกบ้านล่าสุด โดยแต่ละครั้งต้องไม่ต่ำกว่า 1,000 และไม่เกิน 5,000 (ผู้ใช้ต้องใส่อยู่ในช่วงนี้) และนำค่าอาหารทั้ง 3 ครั้งมาบวกกัน หากเกิน 6,000 บาท จะได้รับส่วนลด 15% หากเกิน 12,000 จะได้รับส่วนลด 25% จงหายอดเงินสุดท้ายที่ต้องจ่าย


```python
value1 = int(input("กรุณาป้อนค่าอาหารครั้งที่ 1 (อยู่ระหว่าง 1000 - 5000): "))
value2 = int(input("กรุณาป้อนค่าอาหารครั้งที่ 2 (อยู่ระหว่าง 1000 - 5000): "))
value3 = int(input("กรุณาป้อนค่าอาหารครั้งที่ 3 (อยู่ระหว่าง 1000 - 5000): "))

lst = [value1 ,value2 ,value3]
total = sum(lst)

if (total > 6000):
    net_total = total * (1 - 0.15)
elif (total > 12000):
    net_total = total * (1 - 0.25)
else:
    net_total = total

print("ยอดเงินสุดท้ายที่ต้องจ่าย = " + str(net_total))
```

```python
lst = []
count = 0

while len(lst) < 4:
    value = int(input("กรุณาป้อนค่าอาหารครั้งที่ {} (อยู่ระหว่าง 1000 - 5000): ".format(count+1)))
    if value >= 1000 and value <= 5000:
        lst.append(value)
        count += 1
    else:
        value = int(input("กรุณาป้อนค่าอาหารให้อยู่ระหว่าง 1000 - 5000: "))
        if value >= 1000 and value <= 5000:
            lst.append(value)
            count += 1

total = sum(lst[0:3])
if (total > 4000):
    price_4 = lst[3] * (1 - 0.25)
elif (total > 9000):
    price_4 = lst[3] * (1 - 0.30)
else:
    price_4 = lst[3]
    
print("หายอดเงินสุดท้ายที่ต้องจ่ายของมื้อที่ 4 = " + str(price_4))
```

10. เขียนโปรแกรมให้รับค่าอาหาร 4 ครั้งที่ไปทานอาหารนอกบ้านล่าสุด โดยแต่ละครั้งต้องไม่ต่ำกว่า 1,000 และไม่เกิน 5,000 (เขียนให้ผู้ใช้ใส่ใหม่หากไม่อยู่ในช่วงที่กำหนด) และถามผู้ใช้มีบัตรเครดิตหรือไม่(True/False) นำค่าอาหาร 3 ครั้งแรกมาบวกกัน หากเกิน 4,000 บาท จะได้รับส่วนลด 25% หากเกิน 9,000 จะได้รับส่วนลด 30% ของมื้อที่ 4 หากมีบัตรเครดิต ลดเพิ่มอีก 5%    จงหายอดเงินสุดท้ายที่ต้องจ่ายของมื้อที่ 4


```python
lst = []
count = 0

while len(lst) < 4:
    value = int(input("กรุณาป้อนค่าอาหารครั้งที่ {} (อยู่ระหว่าง 1000 - 5000): ".format(count+1)))
    if value >= 1000 and value <= 5000:
        lst.append(value)
        count += 1
    else:
        value = int(input("กรุณาป้อนค่าอาหารให้อยู่ระหว่าง 1000 - 5000: "))
        if value >= 1000 and value <= 5000:
            lst.append(value)
            count += 1


value2 = bool(input("มีบัตรเครดิตหรือไม่(True/False): "))

total = sum(lst[0:3])
if (total > 4000):
    price_4 = lst[3] * (1 - 0.25)
elif (total > 9000):
    price_4 = lst[3] * (1 - 0.30)
else:
    price_4 = lst[3]
    
if value2 == True:
    price_4 = price_4 * (1 - 0.05)
    
print("หายอดเงินสุดท้ายที่ต้องจ่ายของมื้อที่ 4 = " + str(price_4))
```

```python
lst = []
ordered_lst = []
for i in range(4):
    value = input("กรุณาป้อน str: ")
    lst.append(value)

for n in range(4):
    for x in range(4):
        if len(lst[x]) > len(lst[n]):
            lst[x], lst[n] = lst[n], lst[x]
for z in range(-1, -5, -1):
    print(lst[z])
```
